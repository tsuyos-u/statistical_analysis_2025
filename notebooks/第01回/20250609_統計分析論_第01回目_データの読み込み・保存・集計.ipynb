```python
file_path = "/content/drive/MyDrive/Colab Notebooks/2025_統計分析論/20250609_統計分析論_第01回/data/20250609_mba.csv"

# dir_path = Path("/content/drive/MyDrive/Colab Notebooks/2025_統計分析論/20250609_統計分析論_第01回/data/")
# file_name = "20250609_mba.csv"
# file_path = dir_path / file_name

# csvファイルをdataframeとして読み取る
df = pd.read_csv(file_path)
display(df.head())

# csvファイルを保存する
# 注意するべき点は，dataframeのindexも保存されるときに注意
file_path = "/content/drive/MyDrive/Colab Notebooks/2025_統計分析論/20250609_統計分析論_第01回/data/test.csv"
df.to_csv(file_path)

# 保存したcsv ファイルを再び読み込む
df2 = pd.read_csv(file_path)

# Unnamed: 0 という列が追加されてしまう.
display(df2.head())

# Unnamed: 0 は邪魔だから削除し, 今度はto_csv するときにindex=Falseを足す
df2 = df2.drop("Unnamed: 0", axis=1)
df2.to_csv(file_path, index=False)

# 再びcsvファイルを読み込む
df3 = pd.read_csv(file_path)

# 今度はindexが保存される
display(df3.head())

# gender をindexとして取り込む
df4 = df3.set_index('gender')
display(df4.head())
display(df4.loc["女性"].head())

# この状態でcsvファイルを保存する. このとき, index=Falseをつけると
# 当然，indexの情報が失われる．つまり, genderの列が消える
df4.to_csv(file_path, index=False)
df5 = pd.read_csv(file_path)
display(df5.head())

# 自分でindexを指定している状態でcsvを保存するときは，
# index=Falseをつけてはいけない
df4.to_csv(file_path)
df6 = pd.read_csv(file_path)
display(df6.head())
```

```python
display(df.shape)
display(df['gpa'].describe())
```

```python
# groupby
df.groupby(by='gender')['gpa'].mean()

# gender, race ごとに集計する
display(df.groupby(by=['gender', 'race'])['gpa'].mean())
display(df.groupby(by=['gender', 'race'])['gpa'].count())

# 性別・人種・業種ごとにGPAを集計
res = df.groupby(by=['gender', 'race', 'work_industry']).application_id.count()
display(res)
display(res.loc["男性"].head(50))


# admissionとGPAの関係について見てみたい
display(df.admission.unique())
df.loc[:, 'binary'] = df.admission.replace(
    {"入学": 1, "不合格": 0, "補欠合格": np.nan}
)
display(df)

# 相関係数を計算する
display(df[["gpa", "binary"]].corr())
```

```python
file_path = "/content/drive/MyDrive/Colab Notebooks/2025_統計分析論/20250609_統計分析論_第01回/data/close_price.csv"

close = pd.read_csv(file_path, parse_dates=['date'])
display(close)

close = close.set_index('date')

# return に変換
close_to_close = close.sort_index().pct_change(limit=1)
display(close_to_close.loc['2020-03'])
display(close_to_close.loc['2021'])

# 月ごとのリターンを集計するには？
display(close_to_close.groupby(by=close_to_close.index.month).mean() * 100)
display(close_to_close.groupby(by=close_to_close.index.year).mean() * 100)

# close_to_close.loc[:, 'month'] = close_to_close.index.month
# close_to_close.groupby(by='month').mean() * 100
```
